```python
import pandas as pd

data = pd.read_csv('train.csv')
data
```

```python
# Средний возраст пассажиров
data['Age'].mean()
```

```python
# Средний возраст пассажиров по классам
data.groupby(['Pclass'])['Age'].mean()
```

```python
# Распределение пассажиров по полу
data.groupby(['Sex']).size()
```

```python
# В процентах от общего количества
plot = data.groupby(['Sex']).size().plot(kind = 'pie', autopct='%1.0f%%')
```

```python
# Количество выживших
data['Survived'].sum()
```

```python
# Соотношение выживших и погибших
plot = data['Survived'].value_counts().plot(kind = 'pie', autopct='%1.0f%%')
```

```python
survived = data[data['Survived'] == 1]
plot = survived['Pclass'].value_counts().plot(kind = 'pie', autopct='%1.0f%%')
# Выжившие пассажиры по классам
```

```python
# Выжившие по полу
plot = survived['Sex'].value_counts().plot(kind = 'pie', autopct='%1.0f%%')
```

```python
# Средний возраст выживших
survived['Age'].mean()
```

```python
# Средний возраст выживших по классам
survived.groupby(['Pclass'])['Age'].mean()
```

```python
plot = survived.groupby(['Pclass'])['Age'].mean().plot(kind = 'bar')
```

```python
# Для отдельных класов количество выживших по полу
surv_by_class = survived.groupby(['Pclass', 'Sex']).size()
surv_by_class
```

```python
plot = surv_by_class.plot(kind = 'bar')
```
